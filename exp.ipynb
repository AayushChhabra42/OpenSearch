```python
#Creating and defining the index
index_name = 'movies'
index_body = {
    "settings": {
        "index": {
            "knn": True  # Enable k-NN for this index
        }
    },
    "mappings": {
        "properties": {
            "Series_Title": {"type": "text"},
            "Released_Year": {"type": "integer"},
            "Certificate": {"type": "keyword"},
            "Runtime": {"type": "keyword"},
            "Genre": {"type": "keyword"},
            "IMDB_Rating": {"type": "float"},
            "Overview": {"type": "text"},
            "Meta_score": {"type": "float"},
            "Director": {"type": "text"},
            "Star1": {"type": "text"},
            "Star2": {"type": "text"},
            "Star3": {"type": "text"},
            "Star4": {"type": "text"},
            "No_of_Votes": {"type": "integer"},
            "Gross": {"type": "text"},  # Treat as text to handle commas
            "embedding": {
                "type": "knn_vector",
                "dimension": 384  # Adjust based on the embedding dimension
            }
        }
    }
}

# Create the index
if not client.indices.exists(index=index_name):
    client.indices.create(index=index_name, body=index_body)
```

```python
# Function to delete the index

# Check if the index exists and delete it if it does
def del_index(index_name):
    if client.indices.exists(index=index_name):
        client.indices.delete(index=index_name)
        print(f"Index '{index_name}' deleted successfully.")
    else:
        print(f"Index '{index_name}' does not exist.")
```

```python
#ingesting data into the index
from opensearchpy.helpers import bulk

def bulk_index_movies(movies, index_name='movies'):
    actions = [
        {
            "_index": index_name,
            "_id": movie['id'],
            "_source": movie
        }
        for movie in movies
    ]
    success, _ = bulk(client, actions)
    print(f"Successfully indexed {success} movies.")

# Call the function to index movies
bulk_index_movies(movies)
```

```python
response = client.search(index="movies", body={"query": {"match_all": {}}}, size=10)
for hit in response["hits"]["hits"]:
    print(hit["_source"])
```

### Importing the LLM

We will use the LLM to generate embeddings which can be used for semantic searching in our context.We are using miniLM(sentence-tokenizer mapping sentences & paragraphs to a 384 dimensional dense) here which outputs embeddings.
Once the embeddings for the DB have been generated,we can use the LLM

```python
from transformers import AutoTokenizer, AutoModel
import torch


model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)


def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=128)
    with torch.no_grad():
        embeddings = model(**inputs).last_hidden_state.mean(dim=1)
    return embeddings.squeeze().tolist()  
```

```python
from opensearchpy import helpers

def bulk_update_embeddings(index_name, movies):
    actions = []
    for movie in movies:
        # Generate an embedding for the movie description
        embedding = get_embedding(f"""
            Name:{movie["Series_Title"]}
            Description:{movie["Overview"]}
            Genre:{movie["Genre"]}
            Director:{movie["Director"]}
            Stars:{movie["Star1"]}, {movie["Star2"]}, {movie["Star3"]}, {movie["Star4"]}
        """)

        # Prepare the update action
        action = {
            "_op_type": "update",
            "_index": index_name,
            "_id": movie["id"],  # Ensure this ID matches the one used in indexing
            "doc": {"embedding": embedding}
        }
        actions.append(action)

    helpers.bulk(client, actions)


bulk_update_embeddings("movies", movies)
```

```python
def search_movies(query, index_name, top_k=5):
    # Generate embedding for the query
    query_embedding = get_embedding(query)

    # Construct the semantic search query
    search_query = {
        "size": top_k,
        "query": {
            "knn": {
                "embedding": {
                    "vector": query_embedding,
                    "k": top_k
                }
            }
        }
    }

    # Perform the search
    response = client.search(index=index_name, body=search_query)
    return response["hits"]["hits"]

# Example search
results = search_movies("prison break", "movies",top_k=10)
print(results)
```
